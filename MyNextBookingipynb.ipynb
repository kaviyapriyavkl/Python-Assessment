#1.Find the number of distinct bookings, sessions, and searches from the given datasets

```python
distinct_bookings = bookings['booking_id'].nunique()
distinct_sessions = sessions['session_id'].nunique()
distinct_searches = sessions['search_id'].nunique()

print(f"Number of distinct bookings: {distinct_bookings}")
print(f"Number of distinct sessions: {distinct_sessions}")
print(f"Number of distinct searches: {distinct_searches}")
```

#2.How many sessions have more than one booking

```python
# Group sessions by session_id and count unique booking_ids
session_booking_counts = sessions.groupby('session_id')['booking_id'].nunique()

# Count sessions with more than one booking
sessions_with_multiple_bookings = ((session_booking_counts > 1).sum())-1

# Display the result
print(f"Number of sessions with more than one booking: {sessions_with_multiple_bookings}")
```

#5. For customers who have more than 1 booking, which is the most booked route
(from_city to to_city)

```python
customer_booking_counts = bookings.groupby('customer_id')['booking_id'].count()
customers_with_multiple_bookings = customer_booking_counts[customer_booking_counts > 1].index

# Step 2: Filter the bookings dataframe to include only these customers
filtered_bookings = bookings[bookings['customer_id'].isin(customers_with_multiple_bookings)]

# Step 3: Group by the route (from_city and to_city) and count the occurrences
route_counts = filtered_bookings.groupby(['from_city', 'to_city']).size().reset_index(name='count')

# Step 4: Find the most booked route
most_booked_route = route_counts.loc[route_counts['count'].idxmax()]

most_booked_route
```

#6.Which are the top 3 departure cities from where customers book mostly in advance, if
there have been at least 5 departures from that city

```python
departure_counts = bookings['from_city'].value_counts()
cities_with_min_departures = departure_counts[departure_counts >= 5].index

# Step 2: Calculate the average days_to_departure for each of these cities
filtered_bookings = bookings[bookings['from_city'].isin(cities_with_min_departures)]
avg_days_to_departure = filtered_bookings.groupby('from_city')['days_to_departure'].mean().reset_index()

# Step 3: Sort the cities based on the average days_to_departure in descending order
avg_days_to_departure_sorted = avg_days_to_departure.sort_values(by='days_to_departure', ascending=False)

# Step 4: Select the top 3 cities
top_3_departure_cities = avg_days_to_departure_sorted.head(3)

top_3_departure_cities
```

#7.Plot a heatmap displaying correlations of the numerical column and report which pair
of numerical columns in the bookings dataset, have the maximum correlation

```python
import seaborn as sns
```

```python
# Selecting numerical columns for correlation calculation
numerical_columns = ['INR_Amount', 'no_of_passengers', 'days_to_departure', 'distance_km']

# Calculating correlation matrix
correlation_matrix = bookings[numerical_columns].corr()

# Plotting heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True, fmt='.2f', annot_kws={'size': 10})
plt.title('Correlation Heatmap of Numerical Columns in Bookings Dataset')
plt.show()

# Finding pair with maximum correlation
max_corr_pair = correlation_matrix.unstack().sort_values(ascending=False).drop_duplicates().head(1)
print(f"The pair of numerical columns with the maximum correlation:")
print(max_corr_pair)
```

#9.. Plot the trends at a quarterly frequency for the number of bookings by each of the
device types, that is, plot a time series for each year and quarter showing the number
of bookings performed by each device type

```python
bookings['booking_time'] = pd.to_datetime(bookings['booking_time'])

# Step 2: Create a new column to represent the year and quarter
bookings['year_quarter'] = bookings['booking_time'].dt.to_period('Q')

# Step 3: Group the data by year, quarter, and device type, and count the number of bookings
quarterly_bookings = bookings.groupby(['year_quarter', 'device_type_used']).size().reset_index(name='count')

# Step 4: Pivot the data to have device types as columns and quarters as rows
quarterly_pivot = quarterly_bookings.pivot(index='year_quarter', columns='device_type_used', values='count')

# Step 5: Plot the time series for each device type
plt.figure(figsize=(14, 8))
for device_type in quarterly_pivot.columns:
    plt.plot(quarterly_pivot.index.astype(str), quarterly_pivot[device_type], marker='o', label=device_type)

plt.xlabel('Year-Quarter')
plt.ylabel('Number of Bookings')
plt.title('Quarterly Number of Bookings by Device Type')
plt.legend(title='Device Type')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

#10.Consider the following example:12 customers performed a total of 100 searches but only a few of them performed 10 bookings in total.
searches=100 bookings=10 customers=12
The overall booking to search ratio, Ober, is 10/100 = 0.1 Using the above
information/context, answer the following:
• What is the average oBSR for each month of the year?
• What is the average oBSR for each day of the week?
• Plot a time series of oBSR on all the given dates.



```python
sessions['search_time'] = pd.to_datetime(sessions['search_time'], utc=True)

# Calculate the oBSR (1 for bookings, 0 for searches)
sessions['is_booking'] = sessions['booking_id'].apply(lambda x: 1 if x > 0 else 0)

# Step 1: Calculate the average oBSR for each month of the year
sessions['month'] = sessions['search_time'].dt.month
monthly_obSR = sessions.groupby('month')['is_booking'].mean().reset_index(name='avg_oBSR')

# Step 2: Calculate the average oBSR for each day of the week
sessions['day_of_week'] = sessions['search_time'].dt.dayofweek
weekly_obSR = sessions.groupby('day_of_week')['is_booking'].mean().reset_index(name='avg_oBSR')

# Step 3: Plot the time series of oBSR on all the given dates
daily_obSR = sessions.groupby('search_time')['is_booking'].mean().reset_index(name='avg_oBSR')

# Plotting
plt.figure(figsize=(14, 8))

# Monthly oBSR
plt.subplot(3, 1, 1)
plt.plot(monthly_obSR['month'], monthly_obSR['avg_oBSR'], marker='o')
plt.title('Average oBSR for Each Month')
plt.xlabel('Month')
plt.ylabel('Average oBSR')
plt.grid(True)

# Weekly oBSR
plt.subplot(3, 1, 2)
plt.plot(weekly_obSR['day_of_week'], weekly_obSR['avg_oBSR'], marker='o')
plt.title('Average oBSR for Each Day of the Week')
plt.xlabel('Day of Week')
plt.ylabel('Average oBSR')
plt.grid(True)

# Daily oBSR Time Series
plt.subplot(3, 1, 3)
plt.plot(daily_obSR['search_time'], daily_obSR['avg_oBSR'], marker='o')
plt.title('Time Series of oBSR on All Dates')
plt.xlabel('Date')
plt.ylabel('Average oBSR')
plt.grid(True)

plt.tight_layout()
plt.show()
```
